```python
from folium.plugins import MiniMap
from branca.element import Figure


latitudEscInfo = 9.970611434005122
longitudEscInfo = -84.12930216397356

acropolis=[37.97177010076214, 23.72547730974366]

popuptext = '<b>Escuela de Inform√°tica de la Universidad Nacional</b>'
my_map = folium.Map(location=[acropolis[0],acropolis[1]],zoom_start=15,tiles='stamenterrain')

folium.Marker(location=[acropolis[0],acropolis[1]],popup=popuptext,icon=folium.Icon(color='orange',icon='ok-sign')).add_to(my_map)

folium.Circle(location=[acropolis[0],acropolis[1]],color='purple',weight=6,fill_color='red',fill_opacity=0.5,tooltip='Heredia').add_to(my_map)

minimap = MiniMap(tile_layer='stamenterrain', position='bottomright')

my_map.add_child(minimap)

my_map.save('folium.html')

my_map
```

```python
fig = Figure(width=1000, height=500)
fig.add_child(my_map)
fig
```
